```python
from sklearn.svm import SVC
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from functools import partial
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_blobs
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters_grid_poly = {
    'C':  np.linspace(1.0, 1.9, 20),
    'degree': np.linspace(1, 5, 20),
    'gamma': ['scale', 'auto'],
    'coef0': np.linspace(0.0, 0.3, 20)
}
```

```python
parameters_grid = {
    'C':  np.linspace(1.0, 1.9, 20),
    'gamma': ['scale', 'auto'],
    'coef0': np.linspace(0.0, 0.3, 20)
}
```

```python
data = pd.read_csv('geyser.csv')
data['class_num'] = data['class'].apply(lambda value: {'P': 1, 'N': -1}[value])
X_data = preprocessing.MinMaxScaler().fit_transform(data[['x', 'y']].values)
Y_target = data['class_num']
```

```python
svc = GridSearchCV(SVC(kernel='linear'), parameters_grid, scoring='f1', cv=10).fit(X_data, Y_target).best_estimator_
```

```python
bf_svc = GridSearchCV(SVC(kernel='rbf'), parameters_grid, scoring='f1', cv=10).fit(X_data, Y_target).best_estimator_
```

```python
poly_svc = GridSearchCV(SVC(kernel='poly'), parameters_grid_poly, scoring='f1', cv=10).fit(X_data, Y_target).best_estimator_
```

```python
print(poly_svc)
print(bf_svc)
print(svc)
```

```python
h = .01  # step size in the mesh
# create a mesh to plot in
x_min, x_max = X_data[:, 0].min(), X_data[:, 0].max()
y_min, y_max = X_data[:, 1].min(), X_data[:, 1].max()
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
titles = ['SVC with linear kernel','SVC with RBF kernel','SVC with polynomial kernel']
 
for i, clf in enumerate((svc, bf_svc, poly_svc)):
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    plt.subplot(2, 2, i + 1)
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
 
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
 
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
 
    # Plot also the training points
    plt.scatter(X_data[:, 0], X_data[:, 1], c=Y_target, cmap=plt.cm.coolwarm)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])

plt.show()
```
