```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing 

from sklearn.decomposition import PCA

from sklearn import metrics
from sklearn.metrics import confusion_matrix

import sys

from Clustering import *
```

```python
data = pd.read_csv('iris.csv')

x = data.iloc[:,:4]
y = data.iloc[:, 4:]

y_labels = LabelEncoder().fit(y).transform(y)
y_labels = [float(label) for label in y_labels]

y_onehot = OneHotEncoder().fit_transform(y).toarray
```

```python
x = PCA(n_components = 4).fit(x).transform(x)
x_picture = PCA(n_components = 2).fit(x).transform(x)
```

```python
centers, labels = find_clusters(x, 3)
```

```python
#Predicted
plt.scatter(x_picture[:, 0], x_picture[:, 1], c = labels, s=40, cmap = 'viridis')
```

```python
#Real
plt.scatter(x_picture[:, 0], x_picture[:, 1], c = y_labels, s=40, cmap = 'viridis')
```

```python
print("ARI Sklearn: "+ "{0:.3f}".format(metrics.adjusted_rand_score(y_labels, labels)))
print("ARI Custom: " + "{0:.3f}".format(adjusted_rand_index(confusion_matrix(y_labels, labels))))
```

```python
print("Silhouette Sklearn: "+ "{0:.3f}".format(silhouette(x, labels, centers)))
print("Silhouette Custom: " + "{0:.3f}".format(metrics.silhouette_score(x, np.array(labels).reshape(-1, 1))))
```

```python
slh = []
ari = []
n_clust = []

for i in range(1, 5):
    
    centers, labels = find_clusters(x, i)
    ari.append(adjusted_rand_index(confusion_matrix(y_labels, labels)))
    slh.append((silhouette(x, labels, centers)))
    n_clust.append(i)
    
plt.plot(n_clust, ari)
plt.plot(n_clust, slh)
```
